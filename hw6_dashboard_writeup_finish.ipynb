```python
import pandas as pd
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
import bqplot
```

```python
df = pd.read_csv('building_inventory.csv')
df.head()
```

```python
df["Agency Code"] = df["Agency Name"].astype('category').cat.codes
df.head()
```

```python
heatmap_df1 = df.pivot_table(values='Square Footage', index=['Congress Dist'], columns='Agency Code', aggfunc='sum')
heatmap_df1
```

```python
lineplot_df = df.pivot_table(values='Square Footage', index=['Congress Dist','Agency Code','Year Acquired'], aggfunc='sum').reset_index()
lineplot_df
```

```python
line_initial_df = lineplot_df[(lineplot_df['Congress Dist']==0) & (lineplot_df['Agency Code']==6)]
line_initial_df
```

```python
mapping_agency = df.groupby(by=['Agency Name']).agg({
    'Agency Code': 'first',
}).reset_index()
mapping_agency
```

```python
mapping_congress = df.groupby(by=['Congressional Full Name']).agg({
    'Congress Dist': 'first',
}).reset_index()
mapping_congress
```

```python
mySelectedLabel = ipywidgets.Label()

# Grid heat map
col_sc = bqplot.ColorScale(scheme="RdPu",  # RdPu, Reds, Blues
                           min=np.nanmin(heatmap_df1), 
                           max=np.nanmax(heatmap_df1))
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')
x_ax = bqplot.Axis(scale = x_sc, 
                   label='Agency Code')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label = 'Congress Dist')
heat_map = bqplot.GridHeatMap(color = heatmap_df1.values,
                              row = heatmap_df1.index,  # congress dist
                              column = heatmap_df1.columns,  # agency code
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'},
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0},
                              unselected_style = {'opacity': 1.0})

# Line plot
x_sc = bqplot.LinearScale()  # cannot be deleted
y_sc = bqplot.LinearScale()  # cannot be deleted
ax_xcl = bqplot.Axis(scale = x_sc, 
                     label='Year Acquired')
ax_ycl = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label='Square Footage')
lines = bqplot.Lines(x = line_initial_df['Year Acquired'],
                     y = line_initial_df['Square Footage'],
                     scales = {'x': x_sc, 'y': y_sc})

# Interactive, logic inside this function is IMPORTANT!
def get_data_value(change):
    if len(change['owner'].selected) == 1:
        i,j = change['owner'].selected[0]
        # print(i, j)
        congress = mapping_congress[mapping_congress['Congress Dist']==i].iloc[0,]['Congressional Full Name']
        agency = mapping_agency[mapping_agency['Agency Code']==j].iloc[0,]['Agency Name']
        mySelectedLabel.value = 'Congressional Full Name: {}... Agency Name: {}'.format(congress, agency)
        try:
            region_mask = (lineplot_df['Congress Dist']==i) & (lineplot_df['Agency Code'] == j) & (lineplot_df['Year Acquired']!=0)
            lines.x = lineplot_df['Year Acquired'][region_mask]
            lines.y = lineplot_df['Square Footage'][region_mask]
        except:
            lines.x = []
            lines.y = []
# The bqplot heatmap can observe select event. As long as plotly object can listen on select event, plotly can replace the bqplot.
heat_map.observe(get_data_value, 'selected')

fig_heatmap = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig_line = bqplot.Figure(marks = [lines], axes = [ax_xcl, ax_ycl])

fig_heatmap.layout.min_width='500px'
fig_line.layout.min_width='500px'

# VBox is vertical, HBox is horizontal
myDashboard = ipywidgets.VBox([mySelectedLabel, ipywidgets.HBox([fig_heatmap,fig_line])])
myDashboard
```

# Write-up

- Any data transformations or rescalings you did for your plot?

First, I added a new column named "Agency Code" to the dataframe, by converting "Agency Name" string values to numerical values.

Second, a pivot table dataset named "heatmap_df1" was created for heat map visualization. The index is "Congress Dist", the columns are "Agency Code", and the cell values are sum of "Square Footage". 

Third, I create another dataframe named "lineplot_df" for the line plot. It has 4 columns: congress, agency, year, and square footage.

There are some other dataframes created such as "line_initial_df", "mapping_congress", and "mapping_agency" to help building the dashboard.

No rescaling was applied to the data.

- How you dealt with NaN's (empty entries) in your dataset?

Just leave NaN as is. Because grid heap map will defaultly set NaN value as a black color cell.

However, since the NaN value in heat map means the combination of Congress and Agency doesn't have any record in the original dataset, my code passes an empty list to x axis data, along with an empty list to y axis data of the line plot object. Then the dashboard can show an empty line plot, without throwing an python run-time error.

In addition, I filtered all year=0 records out in the line plot, let the year axis visualization make sense.

- Any aesthetic choices you made (colors, layout, plot size, label size) and why, and what you'd like to experiment with if you had more time.

I chose "RdPu" as bqplot color scale schema, 500px min width for both grid heat map and line plot.
